You can extend the malloy model

```python
%%malloy_model flights

source: carriers is bigquery.table('malloy-data.faa.carriers') extend {
    primary_key: code
}

source: flights is bigquery.table('malloy-data.faa.flights') extend {
    join_one: carriers with carrier

    measure: flight_count is count()

    #line_chart
    view: carriers_over_time is {
        group_by: dep_month is dep_time.month
        aggregate: flight_count
        group_by: carriers.nickname
        limit: 10000
    }
}
```
